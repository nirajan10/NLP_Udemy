```python
import pandas as pd
from nltk import word_tokenize
```

```python
df = pd.read_csv('datasets/bbc_text_cls.csv')
df.head()
```

```python
current_index = 0
word2idx = {}

for doc in df['text']:
    tokens = word_tokenize(doc)
    for word in tokens:
        if word not in word2idx:
            word2idx[word] = current_index
            current_index += 1
```

```python
word2idx
```

```python
idx2word = {}
for key, value in word2idx.items():
    idx2word[value] = key
```
