## Importing libraries

```python
import numpy as np
import pandas as pd

import nltk
from nltk import word_tokenize
```

```python
nltk.download('punkt')
```

## Importing dataset

```python
df = pd.read_csv('datasets/bbc_text_cls.csv')
```

```python
df.head()
```

```python
df.shape
```

## TF (Count Vectorizer)

```python
current_count = 0
word2idx = {}
tokenized_docs = []
for doc in df['text']:
    tokens = word_tokenize(doc.lower())
    doc_as_int = []
    for word in tokens:
        if word not in word2idx:
            word2idx[word] = current_count
            current_count += 1
        doc_as_int.append(word2idx[word])

    tokenized_docs.append(doc_as_int)
```

```python
word2idx
```

```python
idx2word = [key for key,value in word2idx.items()]
```

```python
word2idx['filed']
```

```python
idx2word[487]
```

```python
N = df.shape[0]
N
```

```python
V = len(word2idx)
V
```

```python
tf = np.zeros((N,V))
for i, doc_as_int in enumerate(tokenized_docs):
    for j in doc_as_int:
        tf[i, j] += 1
```

```python
tf
```

## IDF

```python
doc_freq = np.sum(tf>0, axis=0)
doc_freq
```

```python
idf = np.log(N/doc_freq)
idf
```

## TF-IDF

```python
tf_idf = tf*idf
```

```python
tf_idf
```

```python
i = np.random.choice(N)
row = df.iloc[i]
print("Label:", row['labels'])
print("Text:", row['text'].split("\n", 1)[0])
print("Top 5 terms:")

scores = tf_idf[i]
indices = (-scores).argsort()

for j in indices[:5]:
    print(idx2word[j])
```

### Exercise (hard): use Scipy's csr_matrix instead
### You cannot use X[i, j] += 1 here

```python
from scipy.sparse import csr_matrix
```

```python
# Create a CSR matrix for term frequency
rows = []
cols = []
data = []

for i, doc_as_int in enumerate(tokenized_docs):
    for j in doc_as_int:
        rows.append(i)
        cols.append(j)
        data.append(1)

tf_csr = csr_matrix((data, (rows, cols)), shape=(N, V)).toarray()
```
